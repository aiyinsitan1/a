```python
# 清洗文本
import regex
data['文本内容'] = data['文本内容'].str.replace(r'[【】]','',regex = True)
data['文本内容'] = data['文本内容'].str.replace(r'[\n\t]','',regex = True)
data['文本内容'] = data['文本内容'].str.replace(r'[展开c]','',regex = True)
pd.set_option('display.max_rows', None)  # 显示所有行
# 使用regex替换掉所有emoji  
pattern = r'[\U00010000-\U0010ffff\u200d\u2600-\u27bf\ue000-\uf8ff\ud800-\udfff\uf900-\ufdff\u1f000-\u1f64f\u1f680-\u1f6ff\u2600-\u26ff\u2700-\u27bf\ufe00-\ufe0f]+'  
# 应用正则表达式替换  
data['文本内容'] = data['文本内容'].apply(lambda x: regex.sub(pattern, '', x)) 
data['文本内容']
```

```python
import matplotlib.pyplot as plt
import jieba
from wordcloud import WordCloud
from collections import Counter  

# 读取停用词表
stopwords = set()
with open('停用词表.txt', 'r', encoding='utf-8') as file:
    for line in file:
        stopwords.add(line.strip())

# 对"data"的["文本内容"]一列进行分词
comments = data['文本内容'].astype(str)
words_list = []
for comment in comments:
    words = [w for w in jieba.cut(comment) if w not in stopwords]
    words_list.extend(words)

# 统计词频  
word_counts = Counter(words_list)  
word_counts = Counter({word: count for word, count in word_counts.items() if word not in {' ', '\u200b'}})  
word_counts
```

```python
# 创建词云对象  
wordcloud = WordCloud(  
    font_path=r'C:\Windows\Fonts\simkai.ttf', 
    max_words=200,  # 最大显示词数  
    width=800,  
    height=400,  
    background_color='white'  
).generate_from_frequencies(word_counts)  
  
# 显示词云图  
plt.figure(figsize=(10, 5))  
plt.imshow(wordcloud, interpolation='bilinear')  
plt.axis('off')  # 不显示坐标轴  
plt.show()
```

```python
# 情感分析
import jieba
from snownlp import SnowNLP
comments = data['文本内容']
# 读取停用词表
stopwords = set()
with open('停用词表.txt', 'r', encoding='utf-8') as file:
    for line in file:
        stopwords.add(line.strip())
# 分词并进行情感分析
for comment in comments:
    seg_list = jieba.cut(comment)
    seg_result = " ".join([word for word in seg_list if word not in stopwords])  # 去除停用词
    
    s = SnowNLP(comment)
    sentiment = s.sentiments
    
    print("评论内容：", comment)
    print("分词结果：", seg_result)
    print("情感分析：", sentiment)
    print("\n")
```

```python
# 主题分析
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# 使用CountVectorizer进行文本向量化，加入停用词处理
vectorizer = CountVectorizer(stop_words=stopwords)
X = vectorizer.fit_transform(comments)

# 使用LatentDirichletAllocation进行主题分析
lda = LatentDirichletAllocation(n_components=4, random_state=0)
lda.fit(X)

# 输出主题分析结果
feature_names = vectorizer.get_feature_names()
for topic_idx, topic in enumerate(lda.components_):
    print(f"主题 {topic_idx}:")
    print(" ".join([feature_names[i] for i in topic.argsort()[:-10 - 1:-1]]))
    print("\n")
```

```python
import pandas as pd  
import jieba.analyse  
import networkx as nx  
import matplotlib.pyplot as plt  
  
# 提取关键词  
def extract_keywords(text, topK=2):  
    keywords = jieba.analyse.extract_tags(text, topK=topK, withWeight=False)  
    return keywords  
  
# 创建关键词共现网络  
def create_keyword_network(texts, topK=2):  
    G = nx.Graph()  
    for text in texts:  
        keywords = extract_keywords(text, topK)  
        for keyword in keywords:  
            G.add_node(keyword)  
            for other_keyword in keywords:  
                if keyword != other_keyword:  
                    G.add_edge(keyword, other_keyword)  
    return G  
  
# 绘制网络图  
def draw_network(G):  
    plt.figure(figsize=(10, 8))  # 减小图像大小以减少空白  
    pos = nx.spring_layout(G, k=0.5, iterations=50)  # 调整布局参数  
    nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=100, font_size=10)  
    plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签  
    plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号  
    plt.show()  

# 处理文本内容列  
texts = data['文本内容'].tolist()  
  
# 创建网络  
keyword_network = create_keyword_network(texts)  

# 绘制网络图  
draw_network(keyword_network)
```

```python
import pandas as pd 
import spacy 
# 加载spaCy的中文模型 
nlp = spacy.load("zh_core_web_sm") 
# 定义关系抽取函数 
def extract_relations(text): 
    doc = nlp(text) 
    relations = [] 
    for ent1 in doc.ents: 
        for ent2 in doc.ents: 
            if ent1 != ent2: 
                # 查找两个实体之间是否存在依赖路径 
                has_path = False 
                for token1 in ent1: 
                    for token2 in ent2: 
                        if token2 in token1.head.children or token1 in token2.head.children: 
                            has_path = True 
                            break 
                    if has_path: 
                        break 
                if has_path: 
                    relations.append((ent1.text, '相关', ent2.text))  
    return relations 
# 对'文本内容'列应用关系抽取函数 
data['关系抽取'] = data['文本内容'].apply(extract_relations) 
# 打印结果，查看关系抽取情况 
print(data[['文本内容', '关系抽取']]) 
# 如果需要将抽取的关系保存回CSV文件，可以使用以下代码 
data.to_csv('relations_extracted.csv', index=False)
```

```python
import jieba.posseg as pseg  
import spacy  
from spacy.matcher import Matcher  
  
# 加载spacy的中文模型  
nlp = spacy.load("zh_core_web_sm")  
  
data = pd.read_csv("孙卓高考.csv")
  
# 对文本内容列进行实体识别  
for index, row in data.iterrows():  
    txt = row['文本内容']  
      
    # 使用jieba进行名字实体识别  
    jieba_entities = [word for word, flag in pseg.cut(txt) if flag.startswith('nr')]  
      
    # 使用spacy进行实体识别  
    doc = nlp(txt)  
    spacy_entities = [(ent.text, ent.label_) for ent in doc.ents]  
      
    # 打印jieba识别的名字实体  
    if jieba_entities:  
        print(f"第{index+1}行名字实体（jieba）：{', '.join(jieba_entities)}")  
      
    # 打印spacy识别的实体  
    if spacy_entities:  
        print(f"第{index+1}行实体（spacy）：{', '.join([f'{ent[0]} ({ent[1]})' for ent in spacy_entities])}")  
      
    # 关系抽取:抽取人名和动词之间的关系
    relations = []  
    for token in doc:  
        if token.dep_ == 'nsubj' and token.head.dep_ == 'ROOT' and token.head.pos_ == 'VERB':  
            relations.append((token.text, token.head.text))  
      
    # 打印关系  
    if relations:  
        print(f"第{index+1}行关系：{', '.join([f'{subj} -> {verb}' for subj, verb in relations])}")
```
